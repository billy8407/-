```python
import webbrowser
import requests
from bs4 import BeautifulSoup
from gtts import gTTS

def gh(url):
    resp=requests.get(url)
    if resp.status_code==200:
        resp.encoding='utf-8'
        soup=BeautifulSoup(resp.text,"html.parser")
        scope=soup.select('#tab1')
        scope2=scope[0].select('.taba')
        h=[]
        for line in scope2:
            h.append(line.text)
        return h
def t(text):
    if text != None:
        a=gTTS(text=text,lang='zh-tw')
        a.save("aa.mp3")
if __name__=='__main__':
    
    r=gh('https://www.udn.com/news/index')
    d={}
    i=0
    while(i<len(r)):
        d['time']=r[i][:5]
        d['title']=r[i][5:]
        print("time[{}]: {}".format(d['time'],d['title']))
        i+=1
    x="聯合新聞網語音播報開始,"    
    for line in r:
        x+=","+str(line[5:])
    print(x)   
    t(str(x))
    webbrowser.open('aa.mp3')
        
    
        
    
```
